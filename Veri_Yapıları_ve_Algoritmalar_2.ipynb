## **Date Soyut Veri Tipi Uygulaması**

```python
class Date:

    # --- Kurucu -------------------------------------------------------------
    def __init__(self, month: int, day: int, year: int):

        self.month = month
        self.day = day
        self.year = year

    # --- getter methodlar ----------------------------------------------------
    def get_month(self):
        return self.month

    def get_day(self):
        return self.day

    def get_year(self):
        return self.year

    # --- Haftanın günü (Zeller formülü) (0=Mon ... 6=Sun) --------------------
    def dayOfWeek(self):
        if self.month < 3:
            self.month += 12
            self.year -= 1
        return (
            ((13 * self.month + 3) // 5)
            + self.day
            + self.year
            + self.year // 4
            - self.year // 100
            + self.year // 400
        ) % 7

    # --- Ayın adı ------------------------------------------------------------
    def monthName(self) -> str:
        months = ["January", "February", "March", "April", "May", "June",
                  "July", "August", "September", "October", "November", "December"]
        return months[self.month - 1]


    # --- String temsili -----------------------------------------------------
    def __str__(self):
        return "%02d/%02d/%04d" % (self.month, self.day, self.year)

    def __lt__(self, other: "Date") -> bool:
        return (self.year, self.month, self.day) < (other.year, other.month, other.day)

    def __le__(self, other: "Date") -> bool:
        return (self.year, self.month, self.day) <= (other.year, other.month, other.day)

    # --- Karşılaştırmalar ---------------------------------------------------
    def __eq__(self, other: "Date"):
        return (self.year, self.month, self.day) == (other.year, other.month, other.day)

    # --- Yardımcılar --------------------------------------------------------
    @staticmethod
    def _isLeapYear(year: int) -> bool:
        return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)


```

```python
# Kullanıcıdan bir dizi doğum tarihi alır ve
# her bir kişinin en az 25 yaşında olup olmadığını belirler.

def main():
    # Bir kişinin 25 veya daha büyük yaşta olması için doğmuş olması gereken tarih.
    bornBefore = Date(6, 1, 1999)

    # Kullanıcıdan doğum tarihlerini alır ve 25 veya daha büyük olup olmadıklarını belirler.
    date = promptAndExtractDate()
    while date is not None:
        if date <= bornBefore:
            print("En az 25 yaşında: ", date)
        else:
          print("25 yaşından küçük:", date)
        date = promptAndExtractDate()

# Gregoryen tarih bileşenlerini ister ve alır. Kullanıcı tarih girmeyi bitirdiğinde None (Yok) veya bir Tarih nesnesi döndürür.
def promptAndExtractDate():
    print("Bir doğum tarihi girin.")
    month = int(input("ay (çıkış için 0): "))
    if month == 0:
        return None
    else:
        day = int(input("gün: "))
        year = int(input("yıl: "))
        return Date(month, day, year)

# Ana rutini çağır.
main()
```

# Fonksiyonlar ve bilgi aktarımı

```python
def fun(data, value):

    print("ID of data in fun before is ", id(data))
    print("ID of value in fun before is ", id(value))

    data.append(value)
    value = 0

    print("ID of data in fun after is ", id(data))
    print("ID of value in fun after is ", id(value))

    return value
```

```python
dat_a = [1, 2, 5, 7, 9]
valu_e = 10

print("ID of dat_a before fun is ", id(dat_a))
print("ID of valu_e before fun is ", id(valu_e))

new_val = fun(dat_a, valu_e)
print(new_val)

print("ID of dat_a after fun is ", id(dat_a))
print("ID of valu_e after fun is ", id(valu_e))

```
