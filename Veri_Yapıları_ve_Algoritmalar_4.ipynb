# **Yığın (Stack)**

image.png

LIFO (Last In First Out)

## Yığın yapısını gerçekleştirmek için hangi işlemler gereklidir?

**S.push(e):** Yığın S'nin en üstüne e'yi ekle.

**S.pop()**: Yığın S'nin en üstündeki değeri çıkar ve döndür. Yığın boş ise hata vermei.

**S.top():** Yığın S'nin en üst elemanına bir referans döndür. Yığın boş ise hata vermeli.

**S.is_empty():** Yığın boş ise True döndür.

**len(S):** Yığın S'deki eleman sayısını döndür.

## Örnek bir çalıştırma:

image.png

## Yığın yapısının uygulanması:

```python
class PythonListStack:
  def __init__(self):
    self._data=[]

  def __len__(self):
    return len(self._data)

  def is_empty(self):
    return len(self._data) == 0

  def push(self,e):
    self._data.append(e)

  def top(self):
    if self.is_empty():
      return "Stack is empty"
    return self._data[-1]

  def pop(self):
    if self.is_empty():
      return "Stack is empty"
    return self._data.pop()
```

```python
S = PythonListStack()
S.push(5)
S.push(3)
print(len(S))
print(S.pop())
print(S.is_empty())
print(S.pop())
print(S.is_empty())
print(S.pop())
S.push(7)
S.push(9)
print(S.top())
S.push(4)
print(len(S))
print(S.pop())
S.push(6)
S.push(8)
print(S.pop())
```

## Yığın Yapısı ile Bir Dosya İçeriğini Tersine Çevirme

```python
def reversefile(filename):
  S = PythonListStack()
  original = open(filename)
  for line in original:
    S.push(line.rstrip('\n'))
  original.close()
  output = open(filename,'w')
  while not S.is_empty():
    output.write(S.pop() + '\n')
  output.close()

reversefile('/content/dosya.txt')
```

# **Kuyruk (Queue)**

image.png

FIFO (First In First Out)

## Kuyruk yapısını gerçekleştirmek için hangi işlemler gereklidir?

**Q.enqueue(e):** Kuyruğun sonuna eleman ekle.

**Q.dequeue():** Kuyruktaki ilk elemanı çıkar ve döndür. Kuyruk boş ise hata verir

**Q.first():** Kuyuruğun en başındaki elemana bir referans döndürür. Kuyruk boşsa hata verir.

**Q.is_empty():** Kuyruk boşsa True döndürür.

**len(Q):** Kuyruktaki eleman sayısını döndrürür.

## Örnek bir çalıştırma:

image.png

## Kuyruk yapısının uygulanması:

```python
class Queue:
  def __init__(self):
    self._qList = list()

  def isEmpty(self):
    return len(self)==0

  def __len__(self):
    return len(self._qList)

  def enqueue(self, item):
    self._qList.append(item)

  def dequeue(self):
    assert not self.isEmpty(), "Cannot dequeue from an empty queue."
    return self._qList.pop(0)
```

```python
q = Queue()

q.enqueue(1)
q.enqueue(2)
q.enqueue(3)

print(q.dequeue())
print(q.dequeue())
```

```python
class PythonListQueue:

  DEFAULTCAPACITY=5

  def __init__(self):
    self._data = [None] * PythonListQueue.DEFAULTCAPACITY
    self._size = 0
    self._front = 0

  def __len__(self):
    return self._size

  def is_empty(self):
    return self._size == 0

  def first(self):
    if self.is_empty():
      return "Queue is empty"
    return self._data[self._front]

  def dequeue(self):
    if self.is_empty():
      return "Queue is empty"
    answer = self._data[self._front]
    self._data[self._front] = None
    self._front = (self._front + 1) % len(self._data)
    self._size -= 1
    return answer

  def enqueue(self, e):
    if self._size == len(self._data):
      self._resize(2 * len(self._data))
    avail = (self._front + self._size) % len(self._data)
    self._data[avail] = e
    self._size += 1

  def _resize(self, cap):
    old = self._data
    self._data = [None] * cap
    walk = self._front
    for k in range(self._size):
      self._data[k] = old[walk]
      walk = (1 + walk) % len(old)
      self._front = 0
```

```python
Q = PythonListQueue()
Q.enqueue(5)
Q.enqueue(3)
Q.enqueue(5)
Q.enqueue(3)
Q.enqueue(1)
print(len(Q))
print(Q.dequeue())
print(Q.is_empty())
print(Q.dequeue())
print(Q.is_empty())
print(Q.dequeue())
Q.enqueue(7)
Q.enqueue(9)
print(Q.first())
Q.enqueue(4)
print(len(Q))
print(Q.dequeue())
```

```python
from array import array

class Queue:
  def __init__(self, maxSize):
    self._count = 0
    self._front = 0
    self._back = maxSize - 1
    self._qArray = array('i', [0] * maxSize)

  def isEmpty(self):
    return self._count == 0

  def isFull(self):
    return self._count == len(self._qArray)

  def __len__(self):
    return self._count

  def enqueue(self,item):
    assert not self.isFull(), "Cannot enqueue to a full queue."
    maxSize = len(self._qArray)
    self._back = (self._back + 1) % maxSize
    self._qArray[self._back] = item
    self._count += 1

  def dequeue(self):
    assert not self.isEmpty(), "Cannot dequeue from an empty queue."
    item = self._qArray[self._front]
    maxSize = len(self._qArray)
    self._front = (self._front + 1) % maxSize
    self._count -= 1
    return item

my_q = Queue(5)
my_q.enqueue(5)
my_q.enqueue(3)
print(my_q._qArray)
my_q.enqueue(1)
my_q.enqueue(2)
my_q.enqueue(9)
print(my_q._qArray)
my_q.enqueue(3)
```
