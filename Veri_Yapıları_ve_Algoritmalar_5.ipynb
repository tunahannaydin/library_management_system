## Tek Yönlü Bağlı Listede Arama Yapma

```python
def unorderedSearch(head, target):
  curNode = head
  while curNode is not None:
    if curNode.data == target:
      return True
    curNode = curNode.next
  return False

print(unorderedSearch(sll_head, 18))
```

# **Sıralı Çift Yönlü Bağlı Liste:**

## Sıralı çift yönlü bağlı listede arama:

probe kalıcı bir değişken olarak tutulabilir. Sıralı listede daha verimli arama yapılmasını sağlar.

image.png

```python
def searchDLL(head, tail, probe, target):
  if head is None:      #listenin boş olup olmadığının kontrolü
    return False
  elif probe is None:     #probe None'a düşmüşse yeniden başlat
    probe = head

  if target < probe.data:     #aranan değer probe'dan önceyse
    while probe is not None and target <= probe.data:
      if target == probe.data:
        return True
      else:
        probe = probe.prev
  else:     #aranan değer probe'dan sonraysa
    while probe is not None and target >= probe.data:
      if target == probe.data:
        return True
      else:
        probe = probe.next

  return False      #aranan değer bulunamadı
```

# **LinkedList sınıfının uygulanması:**

```python
class SinglyLinkedList:
  def __init__(self):
    self._head = None
    self._size = 0

  def __len__(self):
    return self._size

  def add(self, item):      #listenin başına ekleme yapar
    newNode = self._SListNode(item)
    newNode.next = self._head
    self._head = newNode
    self._size += 1

  def remove(self, item):
    predNode = None
    curNode = self._head

    while curNode is not None and curNode.item != item:
      predNode = curNode
      curNode = curNode.next

    assert (curNode is not None), "The item must be in the list"

    self._size -= 1
    if curNode is self._head: # Use self._head instead of head
      self._head = curNode.next
    else:
      predNode.next = curNode.next

    return curNode.item

  def __iter__(self):
    return self._SListIterator(self._head)

  class _SListIterator:
    def __init__(self, head):
      self._curNode = head

    def __iter__(self):
      return self

    def __next__(self): # Use __next__ instead of next
      if self._curNode is None:
        raise StopIteration
      else :
        item = self._curNode.item
        self._curNode = self._curNode.next
        return item


  class _SListNode:
    def __init__(self, item):
      self.item = item
      self.next = None
```

```python
sll = SinglyLinkedList()
sll.add(12)
sll.add(74)
sll.add(23)
sll.add(19)

for item in sll:
  print(item)

sll.remove(23)

print("---")
for item in sll:
  print(item)
```
